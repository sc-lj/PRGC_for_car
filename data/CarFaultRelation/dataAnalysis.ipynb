```python
import json
from collections import Counter 
with open("all_train.json",'r') as f:
    lines = f.readlines()

lengths = []
for line in lines:
    line= json.loads(line)
    text = line['text']
    lengths.append(len(text))

print(max(lengths))
print(sorted(Counter(lengths).items(),key=lambda x:x[0],reverse=True))
```
